```python

import sqlite3
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt


class RegressionAnalyzer:
    def __init__(self, training_data_files, ideal_functions_file, test_data_file):
        self.training_data_files = training_data_files
        self.ideal_functions_file = ideal_functions_file
        self.test_data_file = test_data_file
        self.db_connection = None
        self.db_cursor = None

    def load_training_data(self):
        # Create SQLite database
        self.db_connection = sqlite3.connect('regression_data.db')
        self.db_cursor = self.db_connection.cursor()

        # Create training data table
        self.db_cursor.execute('''CREATE TABLE IF NOT EXISTS training_data
                                   (X REAL, Y1 REAL, Y2 REAL, Y3 REAL, Y4 REAL)''')

        # Load training data from CSV files
        for file in self.training_data_files:
            df = pd.read_csv(file)
            values = [(row['x'], row['y1'], row['y2'], row['y3'], row['y4']) for _, row in df.iterrows()]
            self.db_cursor.executemany('INSERT INTO training_data VALUES (?, ?, ?, ?, ?)', values)

        self.db_connection.commit()

    def load_ideal_functions(self):
        # Create ideal functions table
        columns = ['X'] + [f'Y{i}' for i in range(1, 51)]  # Assuming there are 50 Y columns
        placeholders = ', '.join(['?'] * len(columns))
        query = f"CREATE TABLE IF NOT EXISTS ideal_functions ({', '.join(columns)})"
        self.db_cursor.execute(query)

        # Load ideal functions from CSV file
        df = pd.read_csv(self.ideal_functions_file)
        values = [tuple(row) for _, row in df.iterrows()]
        query = f"INSERT INTO ideal_functions VALUES ({placeholders})"
        self.db_cursor.executemany(query, values)

        self.db_connection.commit()

    def least_square_fit(self, x, y):
        # Fit a polynomial curve using least squares method
        popt, _ = curve_fit(lambda x, *coeff: sum(coeff[i] * x ** i for i in range(len(coeff))), x, y)
        return popt.tolist()

    def choose_ideal_functions(self):
        # Get training data from the database
        self.db_cursor.execute('SELECT * FROM training_data')
        training_data = self.db_cursor.fetchall()

        # Get ideal functions from the database
        self.db_cursor.execute('SELECT * FROM ideal_functions')
        ideal_functions = self.db_cursor.fetchall()

        best_fit_functions = []
        for i in range(1, 5):
            x = np.array([row[0] for row in training_data])
            y = np.array([row[i] for row in training_data])

            deviations = []
            for j in range(1, len(ideal_functions[0])):
                ideal_y = np.array([row[j] for row in ideal_functions])
                fit_coeff = self.least_square_fit(x, y)
                fitted_y = sum(fit_coeff[i] * x ** i for i in range(len(fit_coeff)))
                deviation = np.sqrt(np.sum((fitted_y - ideal_y) ** 2))
                deviations.append(deviation)

            best_fit_index = np.argmin(deviations) + 1
            best_fit_functions.append(best_fit_index)

        return best_fit_functions

    def load_test_data(self):
        # Create test data table
        self.db_cursor.execute('''CREATE TABLE IF NOT EXISTS test_data
                                   (X REAL, Y REAL, IdealFunction INTEGER, Deviation REAL)''')

        # Load test data from CSV file
        df = pd.read_csv(self.test_data_file)
        self.db_cursor.execute('SELECT * FROM ideal_functions')
        ideal_functions = self.db_cursor.fetchall()

        for _, row in df.iterrows():
            x = row['X']
            y = row['Y']
            deviations = []
            for j in range(1, len(ideal_functions[0])):
                ideal_y = np.array([row[j] for row in ideal_functions])
                fit_coeff = self.least_square_fit(x, y)
                fitted_y = sum(fit_coeff[i] * x ** i for i in range(len(fit_coeff)))
                deviation = np.sqrt(np.sum((fitted_y - ideal_y) ** 2))
                deviations.append(deviation)

            best_fit_index = np.argmin(deviations) + 1
            deviation = min(deviations)
            self.db_cursor.execute('INSERT INTO test_data VALUES (?, ?, ?, ?)',
                                   (x, y, best_fit_index, deviation))

        self.db_connection.commit()

    def plot_test_data(self):
        # Get test data from the database
        self.db_cursor.execute('SELECT * FROM test_data')
        test_data = self.db_cursor.fetchall()

        x = [row[0] for row in test_data]
        y = [row[1] for row in test_data]
        ideal_function_indices = [row[2] for row in test_data]

        self.db_cursor.execute('SELECT * FROM ideal_functions')
        ideal_functions = self.db_cursor.fetchall()

        ideal_function_lines = []
        for i in range(1, len(ideal_functions[0])):
            ideal_y = [row[i] for row in ideal_functions]
            ideal_function_lines.append(ideal_y)

        plt.scatter(x, y, color='blue', label='Test Data')
        for i, idx in enumerate(ideal_function_indices):
            plt.plot(x, ideal_function_lines[idx-1], color='red', label=f'Ideal Function {idx}')

        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Test Data with Ideal Functions')
        plt.legend()
        plt.show()

    def analyze(self):
        self.load_training_data()
        self.load_ideal_functions()
        best_fit_functions = self.choose_ideal_functions()
        self.load_test_data()
        self.plot_test_data()


# Example usage
training_data_files = ['/content/train.csv']
ideal_functions_file = '/content/ideal.csv'
test_data_file = '/content/test.csv'

analyzer = RegressionAnalyzer(training_data_files, ideal_functions_file, test_data_file)
analyzer.analyze()
```
